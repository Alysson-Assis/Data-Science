# Baixando Libs e puxando dados!!

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader as web

!pip install yfinance
import yfinance as yf
```

```python
ibov = yf.download("^BVSP")

ibov.head()
```

```python
ibov['Close'].plot(figsize=(22,8), label="IBOV")
ibov['Close'].rolling(21).mean().plot(label="MM21")
ibov['Close'].rolling(200).mean().plot(label="MM200")
plt.legend()
```

#Selecionado ano

```python
ibov_2008 = ibov[ibov.index.year == 2008]

ibov_2008['Close'].plot(figsize=(22,8), label="IBOV")
ibov_2008['Close'].rolling(21).mean().plot(label="MM21")
ibov_2008['Close'].rolling(200).mean().plot(label="MM200")
plt.legend()
```

#Teste Lógico

```python
ibov_2008 = ibov[(ibov.index.year >= 2008) & (ibov.index.year <= 2015)]

ibov_2008['Close'].plot(figsize=(22,8), label="IBOV")
ibov_2008['Close'].rolling(21).mean().plot(label="MM21")
ibov_2008['Close'].rolling(200).mean().plot(label="MM200")
plt.legend()
```

#Filtro por Mês

```python
ibov_2008 = ibov[ibov.index.month == 12]
ibov_2008.tail(50)
```

#Filtro por Dia

```python
ibov_2008 = ibov[ibov.index.day == 14]
ibov_2008.tail(50)
```
